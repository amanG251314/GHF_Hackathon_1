## Categorical to Numerical

```python
#training data
cat_col=[col for col in train_data.columns if train_data[col].dtype=='O']
Encoders={}
for col in cat_col:
    Encoders['L_enc_'+str(col)]=LabelEncoder()
    train_data[col]=Encoders['L_enc_'+str(col)].fit_transform(train_data[col])
```

```python
for col in cat_col:
    test_data[col]=Encoders['L_enc_'+str(col)].transform(test_data[col])
```

## X,y

```python
xtrain=train_data.drop(['risk_flag'],axis=1)
ytrain=train_data["risk_flag"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(xtrain,ytrain,test_size=0.3,random_state=2021)
```

```python
# Apply ExtratreeClassifier
#et_model = RandomForestClassifier(oob_score= False, n_estimators= 500,min_samples_split= 4,min_samples_leaf= 9, max_features= 7,max_depth= 20, criterion= 'entropy', bootstrap= True,class_weight={0:1,1:10})
et_model = ExtraTreesClassifier(oob_score= False, n_estimators= 500,min_samples_split= 4,min_samples_leaf= 9, max_features= 7,max_depth= 20, criterion= 'entropy', bootstrap= True,class_weight={0:1,1:10})
et_model.fit(X_train, y_train)
ytrain_pred = et_model.predict_proba(X_train)
print('RF train roc-auc: {}'.format(roc_auc_score(y_train, ytrain_pred[:,1])))
ytest_pred = et_model.predict_proba(X_test)
print('RF test roc-auc: {}'.format(roc_auc_score(y_test, ytest_pred[:,1])))
```

```python
#Apply XGBoostClassifier
xg_model=XGBClassifier(scale_pos_weight=9,min_child_weight=7,max_depth=12,learning_rate= 0.2,gamma=0.3,colsample_bytree=1)
xg_model.fit(X_train, y_train)
ytrain_pred = xg_model.predict_proba(X_train)
print('RF train roc-auc: {}'.format(roc_auc_score(y_train, ytrain_pred[:,1])))
ytest_pred = xg_model.predict_proba(X_test)
print('RF test roc-auc: {}'.format(roc_auc_score(y_test, ytest_pred[:,1])))
```

```python
#Apply CatBoostClassifier
cat_model=CatBoostClassifier(class_weights={0:1, 1:10})
cat_model.fit(X_train, y_train)
ytrain_pred = cat_model.predict_proba(X_train)
print('RF train roc-auc: {}'.format(roc_auc_score(y_train, ytrain_pred[:,1])))
ytest_pred = cat_model.predict_proba(X_test)
print('RF test roc-auc: {}'.format(roc_auc_score(y_test, ytest_pred[:,1])))
```

```python
#Apply RandomForestClassifer
rf_model_weighted=RandomForestClassifier(class_weight={0:1, 1:10})
rf_model_weighted.fit(X_train, y_train)
ytrain_pred = rf_model_weighted.predict_proba(X_train)
print('RF train roc-auc: {}'.format(roc_auc_score(y_train, ytrain_pred[:,1])))
ytest_pred = rf_model_weighted.predict_proba(X_test)
print('RF test roc-auc: {}'.format(roc_auc_score(y_test, ytest_pred[:,1])))
```

```python
test_data.columns=xtrain.columns
```

```python
pred=[]
for model in [et_model,xg_model,cat_model,rf_model_weighted]:
    pred.append(pd.Series(model.predict_proba(test_data)[:,1]))
final_prediction=pd.concat(pred,axis=1).mean(axis=1)
#print('Ensemble test roc-auc: {}'.format(roc_auc_score(y_test,final_prediction)))
```
